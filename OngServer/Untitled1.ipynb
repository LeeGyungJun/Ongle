```python
import urllib
import pandas as pd
import pymysql
from urllib.request import urlopen
from bs4 import BeautifulSoup
 
 
# Open database connection
db = pymysql.connect(host='', port=, user='', passwd='', db='',charset='utf8',autocommit=True)
 
# prepare a cursor object using cursor() method
cursor = db.cursor()
 
# SQL문 실행
sql = "select * from gps where gps_id = '9'"
cursor.execute(sql)
 
# 데이타 Fetch
rows = cursor.fetchall()
user_data = list(rows[0])
gps_data = user_data[3]
gps_data_array = gps_data.split("||")
array = []
for i in range(len(gps_data_array)):
    array.append(gps_data_array[i].split(","))
    
lens = len(array)
Date_array = []
Time_array = []
Day_of_week = []
user_data = []

for i in range(0,lens-1):
    Lat_array.append(array[i][1])
    Lng_array.append(array[i][2])
    # 일렬번호,위도,경도,날짜
    del array[i][0]
    # 년-월-일 | 시간:분:초
    Date_array.append(array[i][2].split(" | "))
    # 년-월-일
    Day_of_week.append(Date_array[i][0].split("-"))
    Time_array.append(Date_array[i][1].split(":"))
    # 위도,경도,날짜
    del array[i][2]
    if (Day_of_week[i][3] == '월'):
        Day_of_week[i][3] = '1'
    elif (Day_of_week[i][3] == '화'):
        Day_of_week[i][3] = '2'
    elif (Day_of_week[i][3] == '수'):
        Day_of_week[i][3] = '3'
    elif (Day_of_week[i][3] == '목'):
        Day_of_week[i][3] = '4'
    elif (Day_of_week[i][3] == '금'):
        Day_of_week[i][3] = '5'
    elif (Day_of_week[i][3] == '토'):
        Day_of_week[i][3] = '6'
    elif (Day_of_week[i][3] == '일'):
        Day_of_week[i][3] = '7'
    user_data.append((array[i][0]+","+array[i][1]+","+Day_of_week[i][3]+","+Time_array[i][0]).split(","))
    
    
dataframe = pd.DataFrame(user_data)
dataframe.to_csv("user_data.csv", header=False, index=False, encoding="utf-8-sig")

print(user_data)

# Connection 닫기
db.close()
```
